```python
a = torch.tensor([5.], requires_grad=True)
b = torch.tensor([6.], requires_grad=True)   #  calculating the gradient of a point automatically
```

```python
y = a**3 - b**2
```

```python
y
```

```python
# dy/da = 3*a**2 = 75
# dy/db = -2b = -12
```
