In Flask, app routing refers to the process of mapping URLs to specific functions in your application. This is done using the @app.route decorator, which allows you to define a URL pattern and associate it with a function that should be executed when that URL is requested.

```python
from flask import Flask

app = Flask(__name__)

@app.route('/')
def index():
    return 'This is the homepage.'

@app.route('/about')
def about():
    return 'This is the about page.'

if __name__ == '__main__':
    app.run()
```

We use app routes in Flask to create a well-structured and organized URL scheme for our web application. This makes it easier for users to navigate our application, and it also makes it easier for search engines to index our content. Additionally, app routing allows us to separate the functionality of our application into different functions or modules, which makes our code easier to read, understand, and maintain.

In Flask, the url_for() function is used for URL building. This function takes the endpoint name and any arguments for the endpoint as keyword arguments, and returns a URL for that endpoint.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return '<h1>Hello, World!</h1>'

@app.route('/user/<username>')
def show_user_profile(username):
    return f'<h1>User: {username}</h1>'

@app.route('/post/<int:post_id>')
def show_post(post_id):
    return f'<h1>Post: {post_id}</h1>'

@app.route('/login')
def login():
    return '<h1>Login Page</h1>'

with app.test_request_context():
    print(url_for('index'))
    print(url_for('show_user_profile', username='John Doe'))
    print(url_for('show_post', post_id=123))
    print(url_for('login'))
```
