```python
marks = int(input('Enter your marks in percentage and know your grades :'))
prefix = 'your grade is '
if marks > 100 or marks < 0:
    print('invalid marks')
elif marks>90:
    print(prefix + 'A')
elif marks<=90 and marks >80:
    print(prefix + 'B')    
elif marks<=80 and marks >=60:
    print(prefix + 'C')
else:
    print(prefix + 'D')    
```

```python
#pattern 1
def triangle(n):
    print('1. Hollow Triangle :')
    i = 1
    while i <= n:
        if i ==1:
            print(" "*(n-i) + '*' )
        elif i==n:
            print('* ' * i)
        else:
            print(" "* (n-i) + '*' + ' '*(2*i-3) + '*')
        i += 1    

#pattern 2           
def LeftPascalTriangle(n):
    print('2. Left Pascal Triangle :')
    i =1
    while i<n:
        print(" "*(n-i) + '*'*i)
        i += 1
    
    i = n 
    while i >= 1:
        print(' '*(n-i) + '*'*i)
        i -= 1

#pattern 3
def Diamond(n):
              print('3. Diamond :')
              i = 1
              while i<n:    
                   print(" "*(n-i) + '* '*i)  
                   i += 1
        
              i = n
              while i>=1:
                  print(" "*(n-i) + '* '*i) 
                  i -= 1
            
        
```

```python
row = int(input('Enter the No. of rows :'))
if row < 1:
    print('invalid row')
    
triangle(row)        
LeftPascalTriangle(row)  
Diamond(row)
```
