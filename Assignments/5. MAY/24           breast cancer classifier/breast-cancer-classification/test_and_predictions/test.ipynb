```python
def load_object(file_path: str) -> object:
            with open(file_path, "rb") as file_obj:
                obj = pickle.load(file_obj)

            return obj
        
        

model = load_object(os.path.join("../" , "artifacts" ,"model.pkl"))
preprocessor = load_object(os.path.join("../" , "artifacts" ,"preprocessor.pkl"))
```

```python
test_df = pd.read_csv("test.csv")

for unwanted_column in ["index" ,"Unnamed: 0" ,"target","_id"]:
    test_df =  test_df.drop(columns=unwanted_column) if unwanted_column in test_df.columns else test_df

test_df
```

```python
transformed_x = preprocessor.transform(test_df)
predictions  = model.predict(transformed_x)
predictions 
```

```python
pred_df = test_df
y_pred = [pred for pred in predictions]

pred_df["target"] = y_pred
target_column_mapping = {0:"Benign", 1:"Malignant"}
pred_df["target"] = pred_df["target"].map(target_column_mapping)

pred_df.to_csv("predicted_file_of_test.csv", index=False)
```
