```python
import numpy as np

# Example: Rolling a fair six-sided die
# Each outcome has an equal probability of 1/6
outcomes = [1, 2, 3, 4, 5, 6]
probabilities = np.ones(len(outcomes)) / len(outcomes)

# Calculate PMF for each outcome
pmf = {outcome: probability for outcome, probability in zip(outcomes, probabilities)}

# Print PMF
for outcome, probability in pmf.items():
    print(f"P(X = {outcome}) = {probability:.3f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Example: Heights of adult males (normal distribution)
mean = 175  # Mean height (cm)
std_dev = 10  # Standard deviation (cm)

# Create a range of heights
x = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 1000)

# Calculate the PDF using the normal distribution formula
pdf = norm.pdf(x, mean, std_dev)

# Plot the PDF
plt.plot(x, pdf, label=f'PDF (μ = {mean}, σ = {std_dev})')
plt.xlabel('Height (cm)')
plt.ylabel('Probability Density')
plt.title('Probability Density Function (PDF)')
plt.legend()
plt.grid(True)
plt.show()
```
