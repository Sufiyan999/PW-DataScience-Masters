```python
import plotly.express as px

# Load the "iris" dataset from Plotly
iris_data = px.data.iris()

# Create a scatter matrix plot using Plotly Express
scatter_matrix = px.scatter_matrix(
    iris_data,
    dimensions=["sepal_width", "sepal_length", "petal_width", "petal_length"],
    color="species",
    title="Scatter Matrix Plot of Iris Dataset",
    labels={col: col.replace('_', ' ').title() for col in iris_data.columns},
    opacity=0.7,
    symbol_sequence=['circle', 'square', 'diamond'],
    category_orders={'species': ['setosa', 'versicolor', 'virginica']}
)

# Update the axis titles
scatter_matrix.update_xaxes(title_font=dict(size=14))
scatter_matrix.update_yaxes(title_font=dict(size=14))

# Update the title font size and position
scatter_matrix.update_layout(
    title_x=0.5,
    title_font=dict(size=16),
    legend_title_text='',
    autosize=False,
    width=1000,
    height=600,
)

# Show the scatter matrix plot
scatter_matrix.show()
```

## ANS_5

A `distplot`, short for distribution plot, is a graphical representation commonly used in statistics and data analysis to visualize the distribution or spread of a dataset. It provides insights into the underlying probability distribution of the data, showing how frequently different values occur and the overall shape of the data.

The main components of a distplot typically include a histogram, which represents the frequency or count of data points falling within specific intervals (bins), and a kernel density estimate (KDE), which is a smoothed curve that provides an approximation of the probability density function of the data.

In simpler terms, a distplot helps you answer questions like:

1. **How are the data values distributed across the range?**
   A distplot allows you to see the concentration or spread of data values. Peaks and valleys in the histogram show where the data is more or less concentrated.

2. **What is the overall shape of the data?**
   The shape of the histogram and KDE can give you clues about the data's distribution: whether it's symmetric (like a bell curve), skewed (lopsided), or multimodal (having multiple peaks).

3. **Where are the data values most likely to be found?**
   The KDE provides a smooth estimate of the data's probability density, giving you an idea of where data values are most likely to occur.

4. **Are there any unusual patterns or outliers?**
   Outliers, or extreme values, may stand out as isolated spikes in the histogram or disrupt the smoothness of the KDE.

Creating a distplot in Python involves using data visualization libraries like `seaborn`, `matplotlib`, or `plotly`. You provide the dataset you want to visualize and customize the appearance of the plot to make it more informative and aesthetically pleasing. Adjusting parameters like bin size, color, transparency, and adding features like rug plots (data point indicators) can enhance the understanding of the data distribution.

In your analysis, make sure to describe the characteristics of the distplot, including the central tendency, spread, symmetry or skewness, and any notable features. Emphasize how the distplot helps reveal patterns in the data and aids in making informed decisions or insights based on its distribution.

```python
#Using Seaborn:
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "tips" dataset from Seaborn
tips_data = sns.load_dataset("tips")

# Create a distplot using Seaborn
sns.displot(tips_data["total_bill"], kde=True, rug=False)

# Set plot labels and title
plt.xlabel("Total Bill Amount")
plt.ylabel("Density")
plt.title("Distribution of Total Bill Amounts")

# Show the plot
plt.show()
```

```python
# Using Matplotlib:
import matplotlib.pyplot as plt
import pandas as pd

# Load the "tips" dataset from Seaborn
tips_data = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")

# Create a distribution plot using Matplotlib
plt.hist(tips_data["total_bill"], bins=20, density=True, alpha=0.6, color='blue')
density_values = tips_data["total_bill"].plot.kde(color ="red")

# Set plot labels and title
plt.xlabel("Total Bill Amount")
plt.ylabel("Density")
plt.title("Distribution Plot of Total Bill Amounts")

# Show the plot
plt.show()
```

```python
# Using Plotly Express:
import plotly.express as px

# Load the "tips" dataset from Plotly
tips_data = px.data.tips()

# Create a distplot using Plotly Express
distplot = px.histogram(
    tips_data,
    x="total_bill",
    marginal="rug",
    title="Distplot of Total Bill",
    labels={"total_bill": "Total Bill"},
     opacity=0.6
)

distplot.update_traces(marker_color="blue")

# Show the distplot
distplot.show()
```

```python
# Using Plotly Express:
import plotly.express as px

# Load the "tips" dataset from Plotly
tips_data = px.data.tips()

# Create a distplot-like visualization using Plotly Express
fig = px.histogram(
    tips_data,
    x="total_bill",
    # nbins=20,
    marginal="box",   # Show box plot on the side
    opacity=0.7,
    title="Distribution of Total Bill Amounts"
)

# Customize the layout
fig.update_layout(
    xaxis_title="Total Bill Amount",
    yaxis_title="Count",
    title_x=0.5,
    legend_title_text="",
    autosize=True,
    width=800,
    height=500,
)

# Show the plot
fig.show()
```
